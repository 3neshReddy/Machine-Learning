```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# create DataFrame
df = pd.DataFrame({
    'Sales': [100, 110, 125, 130, 140, 160, 200, 225, 250, 275, 300, 350,
              375, 400, 420, 450, 475, 500, 525, 550, 600, 625, 650, 700],
    'Month': pd.date_range(start='2019-01', end='2020-12', freq='MS')
})

# set 'Month' as index
df.set_index('Month', inplace=True)

# perform seasonal decomposition
result = seasonal_decompose(df['Sales'], model='multiplicative')

# plot the original data and the decomposed components
plt.figure(figsize=(10, 8))
plt.subplot(4, 1, 1)
plt.plot(df['Sales'])
plt.title('Original Data')
plt.subplot(4, 1, 2)
plt.plot(result.trend)
plt.title('Trend Component')
plt.subplot(4, 1, 3)
plt.plot(result.seasonal)
plt.title('Seasonal Component')
plt.subplot(4, 1, 4)
plt.plot(result.resid)
plt.title('Residual Component')
plt.tight_layout()
plt.show()
```
